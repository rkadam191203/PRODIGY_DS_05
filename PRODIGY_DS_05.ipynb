```python
plt.figure(figsize=(12,6))
sns.heatmap(df.isnull(),cbar=False)
plt.title("Missing Values")
plt.show()
```

```python
df.dropna(
    subset=[
        'Weather_Condition',
        'Severity',
        'Start_Time'
    ],
    inplace=True
)
```

```python
df['Start_Time']=pd.to_datetime(df['Start_Time'])
```

```python
df['Hour']=df['Start_Time'].dt.hour
```

```python
df['Day']=df['Start_Time'].dt.day_name()
```

```python
df['Month']=df['Start_Time'].dt.month_name()
```

```python
plt.figure(figsize=(7,5))

sns.countplot(
    x='Severity',
    data=df
)

plt.title("Accident Severity Distribution")

plt.show()
```

```python
plt.figure(figsize=(12,5))

sns.countplot(
    x='Hour',
    data=df
)

plt.title("Accidents by Hour")

plt.show()
```

```python
plt.figure(figsize=(12,5))

sns.countplot(
    x='Day',
    order=[
        'Monday',
        'Tuesday',
        'Wednesday',
        'Thursday',
        'Friday',
        'Saturday',
        'Sunday'
    ],
    data=df
)

plt.title("Accidents by Day")

plt.show()
```

```python
plt.figure(figsize=(14,6))

df['State'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 States with Accidents")

plt.show()
```

```python
plt.figure(figsize=(15,6))

df['Weather_Condition'].value_counts().head(10).plot(kind='bar')

plt.title("Top Weather Conditions")

plt.show()
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x='Severity',
    y='Temperature(F)',
    data=df
)

plt.title("Temperature vs Severity")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x='Severity',
    y='Visibility(mi)',
    data=df
)

plt.title("Visibility vs Severity")

plt.show()
```

```python
plt.figure(figsize=(10,8))

plt.scatter(
    df['Start_Lng'],
    df['Start_Lat'],
    s=1,
    alpha=0.5
)

plt.title("Accident Hotspots")

plt.xlabel("Longitude")

plt.ylabel("Latitude")

plt.show()

```
